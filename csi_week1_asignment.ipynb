```python
import numpy as np
import pandas as pd
```

```python
#loading the dataset in noteook
df = pd.read_excel(r"C:\Users\anshu\OneDrive\Pictures\Documents\Combined_dataset1.csv.xlsx")
```

```python
#exploring the dataset 
#1 
print(df.head())


#2 shape of dataset
print(f"Dataset Shape (Rows, Columns): {df.shape}")

#3 tail of the dataset(last 5 rows)
print("--- LAST 5 ROWS ---")
display(df.tail())

#4 summary of the dataset
print(" DATASET  SUMMARY")
print(df.info())


```

```python
# handling the the dataset
print("Blanks per column before handling:")
print(df.isnull().sum())
print("-" * 50)

# Create a copy
df_cleaned = df.copy()

#handing the missing values
df_cleaned['discount'] = df_cleaned['discount'].fillna(0)
df_cleaned['seller_name'] = df_cleaned['seller_name'].fillna('Unknown Seller')
df_cleaned['what_customers_said'] = df_cleaned['what_customers_said'].fillna('No Review Summary')
df_cleaned['videos'] = df_cleaned['videos'].fillna('No Video Available')
df_cleaned['seller_information'] = df_cleaned['seller_information'].fillna('No Seller Info')
df_cleaned['variations'] = df_cleaned['variations'].fillna('Standard Variant')

print("Missing values handled -Remaining null data", df_cleaned.isnull().sum().sum())
```

```python
# -> step 4 performing operations

#1
print("SELECTING COLUMNS")
display(df_cleaned[['title', 'initial_price', 'rating', 'category']].head())


#2
print("RATING GREATER THAN 3.0")
filtered_rating_df = df_cleaned[df_cleaned['rating'] > 3.0]
display(filtered_rating_df[['title', 'rating', 'final_price']].head())

print(filtered_rating_df.info())




#3 unique catagorie
print("ALL UNIQUE CATEGORIES IN THE DATASET:")
print(filtered_rating_df['category'].unique())


#4 highest prices
print("TOP 5 MOST EXPENSIVE PRODUCTS (SORTED BY PRICE)")
sorted_price_df = filtered_rating_df.sort_values(by='initial_price', ascending=False)
display(sorted_price_df[['title', 'initial_price', 'rating', 'category']].head())



#5 SAMPLES FROM THE MIDDLE OF THE DATASET
print("RANDOM SAMPLE OF 5 ROWS")
display(filtered_rating_df[['title', 'initial_price', 'rating']].sample(5))
```

```python
# -> STEP  5 REMOVE DUPLICATES

# 1. Count the duplicate rows 
duplicate_count = filtered_rating_df.duplicated().sum()

# 2.
if duplicate_count == 0:
    print("No duplicate data found.")
else:
    print(f"removing the duplicate rows : duplicate found ->{duplicate_count}")
    filtered_rating_df = filtered_rating_df.drop_duplicates()
    print(f"Deduplication complete. Current shape: {filtered_rating_df.shape}")
```

```python
# --- STEP 6: CREATE THE DERIVED COLUMN ---

# Calculate the total_amount directly onto your filtered data rows
filtered_rating_df['total_amount'] = filtered_rating_df['initial_price'] * (filtered_rating_df['discount'] / 100)

# Verify that the calculation column has been added perfectly
display(filtered_rating_df[['title', 'initial_price', 'discount', 'total_amount']].head())
print(filtered_rating_df.info())
```
